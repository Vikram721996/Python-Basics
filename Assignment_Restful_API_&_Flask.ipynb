**Question 1 : What is a RESTful API?**

**Answer:** A RESTful API is an interface that allows applications to interact with each other using standard HTTP methods such as GET, POST, PUT, and DELETE. These methods correspond to the fundamental operations of Read, Create, Update, and Delete (CRUD) on resources.

**Question 2 : What is Flask, and why is it popular for building APIs?**

**Answer:** Flask is a lightweight and flexible Python web framework designed to build web applications and APIs with simplicity and minimalism and it is popular for building APIs due to its lightweight design, flexibility, and ease of use, allowing developers to quickly create and scale RESTful services.

**Question 3 : What are HTTP methods used in RESTful APIs?**

**Answer:** **HTTP Methods in RESTful APIs**

* **GET** - Retrieve data

* **POST**- Create new data

* **PUT** - Replace existing data entirely

* **PATCH** - Update part of existing data

* **DELETE** - Remove data

* **OPTIONS** - Get allowed communication methods for a resource

* **HEAD** - Retrieve headers only (no response body)



**Question 4 : What is the purpose of the @app.route() decorator in Flask?**

**Answer:**  The @app.route() decorator in Flask is used to bind a URL path to a specific function, defining the route that the application responds to when a client makes a request.

**Question 5 : What is the role of Flask-SQLAlchemy?**

**Answer** Flask-SQLAlchemy is an extension for Flask that integrates SQLAlchemy (a popular ORM) into Flask applications. It simplifies database interactions by providing a higher-level abstraction to manage database connections, queries, and models within Flask, making it easier to work with relational databases.

**Question 7 : How do you return JSON responses in Flask?**


**Answer:** Flask provides built-in ways to send responses in JSON format, which is the standard format used in APIs.

* **Using Flask’s built-in JSON response function**

Flask has a helper function that converts Python data (like dictionaries or lists) into a proper JSON response.
It automatically:

* Converts the data into JSON format

* Sets the correct response headers

* Ensures the response is safe and properly formatted

* Adding a status code

You can send a JSON response along with an HTTP status code (such as 200, 400, or 404) so the client knows whether the request succeeded or failed.

* **Creating a custom JSON response**

If you need extra control—such as adding custom headers—you can create a response object manually and attach JSON data to it.
This allows:

* Custom headers

* Custom status

* More flexible formatting

* **Sending raw JSON**

As a last option, you can manually convert your Python data to JSON and return it yourself.
This method is only used when you need full control over the response.

**Question 8 : How do you handle POST requests in Flask?**

**Answer:** POST requests are defined by specifying methods=['POST'] in the route decorator. Inside the function, data sent by the client can be accessed using request.form for form data or request.json for JSON data. The function then returns a response, such as a confirmation message or a status code like 201 for successful creation.

**Q9. How do you handle errors in Flask (e.g., 404)?**

**Answer:** 404 error in Flask means the requested page or route does not exist.

Flask allows you to create a custom 404 error handler, so instead of showing the default error page, you can return your own message, HTML, or JSON. When a user tries to access an invalid URL, Flask automatically triggers this handler and sends your custom response.

**Question 10: How do you structure a Flask app using Blueprints?**

**Answer:** Blueprints in Flask are used to organize your application into modular components.
Instead of putting all routes, logic, and views in one big file, you split them into separate, manageable sections.

**Why Use Blueprints?**

* Keeps your project organized as it grows

* Separates features into modules (e.g., auth, users, admin)

* Allows team members to work independently

* Makes your application scalable and cleaner

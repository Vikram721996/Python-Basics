**Q1. What is hypothesis testing in statistics?**

Hypothesis testing is a statistical method to determine if a sample provides enough evidence to reject a **null hypothesis (H₀)** in favor of an **alternative hypothesis (H₁)**. A **P-value** helps measure the likelihood of observing the data under H₀—if **P-value < α (e.g., 0.05)**, we reject H₀. Common tests include **Z-test, T-test, and Chi-square test**, used in fields like science, finance, and medicine

**Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

The **null hypothesis (H₀)** is a statement that assumes **no effect or no difference** in a population—it suggests that any observed variation is due to chance. The **alternative hypothesis (H₁ or Hₐ)**, on the other hand, proposes that **there is an effect or a difference**, meaning the observed data is not just due to random chance.

**Null Hypothesis (H₀)**
* Assumes no effect or relationship
* To be tested and possibly rejected

**Example :-** "The average height of students is 5.5 ft."

**Alternative Hypothesis (H₁)**
* Suggests an effect or relationship exists
* To be supported if H₀ is rejected

**Example :-** "The average height of students is not 5.5 ft."

**Q3. What is the significance level in hypothesis testing, and why is it important?**

The **significance level (α)** in hypothesis testing is the probability of rejecting the **null hypothesis (H₀)** when it is actually true. It represents the threshold for determining whether an observed effect is statistically significant.

**Why is it important?**
1. **Controls Type I Error:** A lower α (e.g., 0.05) reduces the risk of falsely rejecting H₀.

2. **Defines Statistical Rigor:** Researchers set α (commonly 0.05 or 0.01) to ensure results are meaningful.

3. **Guides Decision-Making**: If **P-value < α**, we reject H₀, indicating strong evidence for the alternative hypothesis.

**Q4. What does a P-value represent in hypothesis testing?**

A **P-value** in hypothesis testing represents the probability of obtaining results as extreme as the observed data, assuming the **null hypothesis (H₀)** is true. It helps determine whether the observed effect is statistically significant.

**Key Points About P-Value:**

1. **Low P-value (< 0.05)** → Strong evidence against H₀ → Reject H₀.

2. **High P-value (> 0.05)** → Weak evidence against H₀ → Fail to reject H₀.

3. **It does not prove H₀ or H₁** but indicates the likelihood of observed data under H₀.

**Q5. How do you interpret the P-value in hypothesis testing?**

1. The **P-value** quantifies the probability of obtaining results as extreme as the observed data, assuming the null hypothesis (H₀) is true.

2. A **low P-value** (typically < 0.05) suggests that such extreme results are unlikely under H₀, providing strong evidence to reject it in favor of the **alternative hypothesis (H₁)**.

3. Conversely, a **high P-value** indicates that the observed data is consistent with H₀, meaning there isn’t enough evidence to reject the null hypothesis.

**Q6. What are Type 1 and Type 2 errors in hypothesis testing?**

 In hypothesis testing, **Type I and Type II errors** represent two kinds of mistakes you might make when drawing conclusions from sample data:

1. **Type I Error (False Positive)**: This occurs when you reject the null hypothesis (H₀) even though it is actually true. Essentially, it's like a false alarm—for example, concluding a drug is effective when it really isn’t. The significance level (α) you choose reflects the risk you're willing to take for this error.

2. **Type II Error (False Negative)**: This happens when you fail to reject the null hypothesis even though it is false. In other words, you miss detecting a true effect, such as not recognizing that a drug is effective when it actually is. The probability of a Type II error is denoted by β, and its complement (1 - β) is known as the test's power.

**Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

A **one-tailed test** evaluates a hypothesis with a directional alternative—for example, testing if a parameter is either greater than or less than a specific value—so the entire significance level (α) is placed in one tail of the distribution. This approach is more powerful for detecting an effect in that specific direction but ignores deviations in the opposite direction .

In contrast, a **two-tailed test** examines whether a parameter is simply different from a specified value by considering extreme outcomes in both directions. Here, the significance level is split between both tails, allowing detection of an effect regardless of its direction, which makes this test more conservative.

**Q8. What is the Z-test, and when is it used in hypothesis testing?**

The **Z-test** is used in hypothesis testing primarily to decide whether to reject the null hypothesis by comparing the observed sample statistic against the known population parameter. It is most appropriate when you have a large sample or when the population variance is available, making it simpler to use than the T-test, which is used when the sample size is small or the population standard deviation is unknown.

**Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

1. **Compare the Value to the Average**: Start by finding the difference between your observed value (like a sample mean) and the average (population mean or expected value). This tells you how much higher or lower your value is compared to what you normally expect.

2. **Understand the Variation**: Next, consider the typical amount of variation in your data—that is, how spread out your values usually are. This variation is what you’d expect from ordinary fluctuations or measurement differences.

3. **Gauge the Unusualness**: Finally, assess how many “steps” of this typical variation fit into the difference you observed. If the difference is small compared to the usual spread, your value is pretty ordinary. If it’s large, then it stands out. That “number of steps” is what the Z-score tells you—essentially, it’s a measure of how unusual your observation is relative to the normal variation in the data.

**Q10. What is the T-distribution, and when should it be used instead of the normal distribution?**

The **T-distribution** is a probability distribution that is similar to the normal distribution but has heavier tails. This means it gives more probability to events farther from the mean, accommodating the extra uncertainty that comes from estimating the population standard deviation using a small sample.

You should use the T-distribution instead of the normal distribution when:

* The sample size is small (commonly n < 30).

* The population standard deviation is unknown—so you estimate it with the sample standard deviation.

* The underlying data is approximately normally distributed.

**Q11. What is the difference between a Z-test and a T-test?**

A **Z-test** is used when you have a large sample size (typically n ≥ 30) or when the population standard deviation is known. In these cases, the sampling distribution of the mean is normally distributed, and the test statistic is based on the standard normal (Z) distribution.

In contrast, a **T-test** is used when the sample size is small (typically n < 30) or when the population standard deviation is unknown. Here, you estimate the standard deviation from your sample, which adds extra uncertainty. The test statistic then follows a t-distribution with degrees of freedom that depend on your sample size, which has heavier tails than the normal distribution to account for this additional variability.

**Q12. What is the T-test, and how is it used in hypothesis testing?**

A **T-test** is a statistical method used to compare means when the population variance is unknown. It is especially useful when the sample size is small. The test calculates a t-statistic by measuring the difference between the sample mean and the hypothesized mean, adjusted by the sample's standard error. This t-statistic is then compared to a t-distribution with degrees of freedom equal to the sample size minus one. If the resulting p-value is below the chosen significance level, the null hypothesis is rejected.

**Q13. What is the relationship between Z-test and T-test in hypothesis testing?**

Both the **Z-test** and **T-test** are used to assess whether a sample mean significantly differs from a hypothesized population mean. The Z-test is appropriate when the population standard deviation is known or when the sample size is large, allowing the Central Limit Theorem to ensure normality. In contrast, the T-test is used when the population standard deviation is unknown and the sample size is small; it compensates for the extra uncertainty by using the sample standard deviation and following a t-distribution. Essentially, the T-test is an adaptation of the Z-test that adjusts for increased variability in small samples. As the sample size grows, the t-distribution increasingly resembles the standard normal distribution, making the two tests converge.

**Q14. What is a confidence interval, and how is it used to interpret statistical results?**

A **confidence interval (CI)** is a range derived from your sample data that estimates where the true population parameter (like a mean or proportion) is likely to fall. It is constructed by taking a point estimate and adding and subtracting a margin of error, which depends on the variability in the data and the chosen confidence level (e.g., 95%). A 95% CI indicates that if you were to take many samples, about 95% of the calculated intervals would contain the true population value. The width of the CI provides insight into the precision of your estimate—a narrower interval suggests higher precision, whereas a wider interval indicates more uncertainty. Confidence intervals help interpret statistical results by showing whether a hypothesized parameter value (such as a null value) falls within a plausible range; if it lies outside, it may signal a statistically significant difference.

**Q15. What is the margin of error, and how does it affect the confidence interval?**

**1.** The margin of error represents the "plus-or-minus" range around a sample estimate, reflecting the potential variability when estimating a population parameter.

**2.** It is calculated by multiplying the standard error by a critical value corresponding to the chosen confidence level.

**3.** When forming a confidence interval, you take your point estimate and add/subtract the margin of error to define the interval's bounds.

**4.** A larger margin of error results in a wider confidence interval, indicating more uncertainty about the true parameter value.

**5.** Conversely, a smaller margin of error produces a narrower interval, suggesting a more precise estimate of the population parameter.

**Q16. How is Bayes' Theorem used in statistics, and what is its significance?**

**Bayes' Theorem** is used in statistics to update the probability of a hypothesis by combining prior beliefs with new evidence. It equates the posterior probability to the product of the likelihood (how probable the new evidence is given the hypothesis) and the prior probability, normalized by the overall probability of the evidence. This systematic updating process makes it invaluable for decision-making under uncertainty, allowing for more flexible and realistic models than methods that ignore prior knowledge. Its significance is also seen in diverse applications—from medical diagnoses to spam detection—where continually revising predictions based on emerging data is crucial. Overall, Bayes' Theorem underpins Bayesian inference, serving as a cornerstone for modern statistical analysis and decision theory.

**Q17. What is the Chi-square distribution, and when is it used?**

The **Chi-square distribution** is a probability distribution that arises when summing the squares of independent standard normal variables. It is primarily used in hypothesis testing, especially for categorical data.

**When is it used?**
1. **Chi-square goodness-of-fit test**: Determines if observed data matches an expected distribution.

2. **Chi-square test for independence**: Assesses whether two categorical variables are related.

3. **Variance estimation**: Used in confidence intervals for population variance when data follows a normal distribution.

**Q18. What is the Chi-square goodness of fit test, and how is it applied?**

The Chi-square goodness of fit test is a statistical method used to determine whether observed categorical data fit a specific expected distribution. It compares the observed frequencies in each category to the expected frequencies calculated from the hypothesized distribution to see how well the data “fits” that distribution.

**How it is applied:**

  **1. Set up hypotheses:**

 * Null hypothesis (H₀): The observed data follows the expected distribution.

 * Alternative hypothesis (H₁): The observed data does not follow the expected distribution.

**Calculate expected frequencies** for each category based on the expected distribution.

**Compute the Chi-square statistic** by measuring the differences between observed and expected frequencies.

**Compare the test statistic** to the Chi-square distribution to decide whether to reject H₀, indicating whether the data fits the expected distribution or not.

**Q19. What is the F-distribution, and when is it used in hypothesis testing?**

The F-distribution is a continuous probability distribution that arises primarily in hypothesis testing and analysis of variance (ANOVA). It describes the ratio of two scaled chi-square distributed variables and is used to compare variances between groups. Specifically, it is used in ANOVA to test if group means are equal and in tests for equality of variances.

**Q20. What is an ANOVA test, and what are its assumptions?**

**ANOVA (Analysis of Variance)** is a statistical test used to determine if there are significant differences between the means of three or more groups.

**The main assumptions of ANOVA are:**

**1. Independence** of observations

**2. Normality —** the data in each group should be approximately normally distributed

**3. Homogeneity of variances —** the variances across groups should be equal

**Q21. What are the different types of ANOVA tests?**

**The main types of ANOVA tests are:**

**1. One-way ANOVA:** Tests differences between means of three or more groups based on one factor.

**2. Two-way ANOVA:** Tests the effect of two factors on the means, including interaction effects.

**3. Repeated Measures ANOVA:** Tests means when the same subjects are measured multiple times under different conditions.

**Q22. What is the F-test, and how does it relate to hypothesis testing?**

The **F-test** is a statistical test that compares the variances of two populations using the ratio of their sample variances. It relates to hypothesis testing by evaluating whether the observed difference in variances is significant enough to reject the null hypothesis that the variances are equal.

**Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the discrete uniform distribution (e.g., a fair six-sided die)
low = 1  # Inclusive lower bound
high = 7  # Exclusive upper bound (so 7 means the range is 1 to 6)
distribution = randint(low, high)

# Generate 10 random samples
samples = distribution.rvs(size=10)
print("Random Samples:", samples)

# Define the range of possible outcomes
x = np.arange(low, high)

# Calculate the PMF for each outcome
pmf_values = distribution.pmf(x)

# Plot the PMF
plt.bar(x, pmf_values, width=0.6, color='skyblue', edgecolor='black')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
from scipy.stats import bernoulli

def bernoulli_pmf(p, k):
    """
    Calculate the PMF of a Bernoulli distribution for a given probability p and outcome k.

    Parameters:
    - p (float): Probability of success (0 <= p <= 1).
    - k (int): Outcome (0 or 1).

    Returns:
    - float: Probability of observing outcome k.
    """
    if k not in [0, 1]:
        raise ValueError("Outcome k must be 0 or 1.")
    return bernoulli.pmf(k, p)

# Example usage:
p = 0.7  # Probability of success
for k in [0, 1]:
    probability = bernoulli_pmf(p, k)
    print(f"P(X = {k}) = {probability}")
```

**Q9. Write a Python function to calculate Z-scores from a dataset and plot them?**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot their histogram.

    Parameters:
    data (array-like): Input dataset (list, numpy array, etc.)

    Returns:
    z_scores (numpy array): Array of Z-scores
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
if __name__ == "__main__":
    # Generate example data (normal distribution)
    sample_data = np.random.normal(loc=50, scale=10, size=1000)
    z_scores = calculate_and_plot_z_scores(sample_data)
```

**Q15. Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [102, 100, 98, 105, 101, 99, 97, 103, 104, 100]

# Known population parameters
pop_mean = 100
pop_std = 5

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate Z-statistic
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print(f"Z-statistic: {z:.3f}")

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")

# Conclusion at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Significant difference found.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

**Q16. Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

# Sample data
data = [12, 15, 14, 10, 13, 16, 12, 14, 15, 11]

# Calculate sample mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean

# Confidence level
confidence = 0.95

# Calculate the confidence interval
ci = stats.t.interval(confidence, len(data)-1, loc=mean, scale=std_err)

print(f"Sample mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

# Interpretation
print(f"We are 95% confident that the true population mean lies between {ci[0]:.2f} and {ci[1]:.2f}.")
```

**Q18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 0
std_dev = 1

# Create x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# Calculate PDF using the formula for normal distribution
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean)/std_dev)**2)

# Plot the PDF
plt.plot(x, pdf, color='red')
plt.title('Normal Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

**Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.**

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample = [102, 100, 98, 105, 101, 99, 97, 103, 104, 100]

# Known population parameters
pop_mean = 100
pop_std = 5

# Sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate Z-statistic
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print(f"Z-statistic: {z:.3f}")

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")

# Interpret result at alpha = 0.05 significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

**Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample = [105, 102, 99, 101, 98, 100, 97, 103, 104, 106]

# Known population parameters
pop_mean = 100
pop_std = 5
sample_size = len(sample)

# Calculate sample mean and Z-statistic
sample_mean = np.mean(sample)
z = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print(f"Z-statistic: {z:.3f}")

# Significance level
alpha = 0.05

# Critical z-values for two-tailed test
z_critical_low = norm.ppf(alpha/2)
z_critical_high = norm.ppf(1 - alpha/2)

print(f"Critical Z-values: {z_critical_low:.3f}, {z_critical_high:.3f}")

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"P-value: {p_value:.4f}")

# Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= z_critical_low), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= z_critical_high), color='red', alpha=0.3)

# Plot the calculated Z-statistic
plt.axvline(z, color='blue', linestyle='--', label=f'Z-statistic = {z:.2f}')

plt.title('Two-tailed Z-test')
plt.xlabel('Z value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Decision
if z < z_critical_low or z > z_critical_high:
    print("Reject the null hypothesis: Significant difference.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

**Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    """
    mu0: mean under null hypothesis
    mu1: mean under alternative hypothesis
    sigma: population standard deviation
    n: sample size
    alpha: significance level (Type 1 error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical z for two-tailed test
    z_crit_low = norm.ppf(alpha / 2)
    z_crit_high = norm.ppf(1 - alpha / 2)

    # Critical values in terms of sample mean
    crit_low = mu0 + z_crit_low * se
    crit_high = mu0 + z_crit_high * se

    # Type 1 error: reject H0 when true mean = mu0 (alpha)
    # Already set by critical values

    # Type 2 error: fail to reject H0 when true mean = mu1
    beta_low = norm.cdf(crit_low, loc=mu1, scale=se)
    beta_high = 1 - norm.cdf(crit_high, loc=mu1, scale=se)
    beta = beta_low + beta_high

    # Plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    plt.plot(x, norm.pdf(x, mu0, se), label='H0 Distribution (mean=mu0)')
    plt.plot(x, norm.pdf(x, mu1, se), label='H1 Distribution (mean=mu1)')

    # Shade Type 1 error regions (tails of H0)
    plt.fill_between(x, 0, norm.pdf(x, mu0, se), where=(x <= crit_low), color='red', alpha=0.3, label='Type 1 Error (α)')
    plt.fill_between(x, 0, norm.pdf(x, mu0, se), where=(x >= crit_high), color='red', alpha=0.3)

    # Shade Type 2 error region (middle of H1)
    plt.fill_between(x, 0, norm.pdf(x, mu1, se), where=((x > crit_low) & (x < crit_high)), color='blue', alpha=0.3, label='Type 2 Error (β)')

    plt.title('Type 1 and Type 2 Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Significance level (Type 1 error, α): {alpha}")
    print(f"Type 2 error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05)
```

**Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper

# Example usage
sample_data = [12, 15, 14, 10, 13, 15, 16, 14, 13, 12]
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

```python
def bayes_theorem(p_disease, p_pos_given_disease, p_no_disease, p_pos_given_no_disease):
    # Total probability of testing positive
    p_pos = p_pos_given_disease * p_disease + p_pos_given_no_disease * p_no_disease

    # Bayes' theorem formula
    p_disease_given_pos = (p_pos_given_disease * p_disease) / p_pos

    return p_disease_given_pos

# Given probabilities
p_disease = 0.01          # Prior probability of disease
p_no_disease = 1 - p_disease
p_pos_given_disease = 0.9  # Sensitivity
p_pos_given_no_disease = 0.05  # False positive rate

posterior = bayes_theorem(p_disease, p_pos_given_disease, p_no_disease, p_pos_given_no_disease)

print(f"Posterior probability of disease given positive test: {posterior:.4f}")
```

**Q12. Perform a Chi-square test for independence between two categorical variables in Python.**

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table (e.g., Gender vs Preference)
# Rows: Male, Female
# Columns: Like, Dislike
data = np.array([[30, 10],
                 [20, 40]])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print(f"Chi-square statistic: {chi2:.3f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject null hypothesis: Variables are dependent (associated).")
else:
    print("Fail to reject null hypothesis: Variables are independent.")
```

**Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.**

```python
import numpy as np

# Observed frequency table (contingency table)
observed = np.array([[30, 10],
                     [20, 40]])

# Calculate row sums, column sums, and total sum
row_sums = observed.sum(axis=1).reshape(-1, 1)
col_sums = observed.sum(axis=0).reshape(1, -1)
total = observed.sum()

# Calculate expected frequencies
expected = (row_sums @ col_sums) / total

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```
